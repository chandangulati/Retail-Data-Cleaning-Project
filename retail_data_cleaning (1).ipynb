```python
import pandas as pd
import numpy as np

data = {
    'Transaction_ID':[1001,1002,1003,1004,1004,1005],
    'Customer_Name':['Rahul','Priya',None,'Amit','Amit','Neha'],
    'Product':['Laptop','Mouse','Keyboard','Monitor','Monitor','Headphones'],
    'Quantity':[1,2,3,1,1,None],
    'Price':[55000,None,1200,15000,15000,2500]
}

df = pd.DataFrame(data)

print("Original Data")
print(df)

df.fillna({
    'Customer_Name':'Unknown',
    'Quantity':df['Quantity'].median(),
    'Price':df['Price'].median()
}, inplace=True)

df.drop_duplicates(inplace=True)

df['Total_Sales'] = df['Quantity'] * df['Price']

df.to_csv('cleaned_retail_transactions.csv', index=False)

print("\nCleaned Data")
print(df)

print("\nCSV Exported Successfully")
```
